16. What is an Optimizer? What are different types of optimizers? Explain each with an example.
 - An **Optimizer** is an algorithm that adjusts model parameters to minimize the loss function during training.

**Types of optimizers:**

I. **Gradient Descent:** Updates parameters using the average gradient of all data (e.g., batch gradient descent).
II. **Stochastic Gradient Descent (SGD):** Updates parameters using the gradient of a single data point, faster but noisier.
III. **Adam:** Combines momentum and adaptive learning rates for faster, efficient training (widely used).
IV. **RMSprop:** Adapts learning rate based on recent gradients, good for non-stationary problems.
